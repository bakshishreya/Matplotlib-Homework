```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merge_df = pd.merge(mouse_metadata, study_results, on="Mouse ID")


# Display the data table for preview
merge_df.head()
```

```python
# Checking the number of mice.
mice_count = merge_df["Mouse ID"].count()
mice_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

duplicate_rows = merge_df[merge_df.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_rows
```

```python
# Optional: Get all the data for the duplicate mouse ID. 

duplicate_rows_all = merge_df[merge_df.duplicated(['Mouse ID'])]
duplicate_rows_all
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
 # Identify incomplete rows
clean_df = merge_df.drop_duplicates("Mouse ID")
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
mice_count_clean = clean_df["Mouse ID"].count()
mice_count_clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
mean=merge_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median=merge_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance=merge_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
standard_deviation=merge_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
Sem=merge_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

summary_df = pd.DataFrame({"Mean": mean,"Median": median,"Variance": variance,"Standard Deviation": standard_deviation,"SEM": Sem})
summary_df
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pandas.
new_data_pybar=merge_df.groupby(["Drug Regimen"]).count()["Mouse ID"]
new_data_pybar.plot(kind="bar", figsize=(10,5))
# Set a title for the chart
plt.ylabel("Number of unique Mice Tested")
plt.title("")
plt.tight_layout()
plt.show()

```

```python
# Generate a bar plot showing the total number of measurements taken on each drug regimen using pyplot.
new_data_pyplot=(merge_df.groupby(["Drug Regimen"])["Mouse ID"].count()).tolist()
x_axis = np.arange(len(new_data_pyplot))
plt.figure(figsize=(10,5))
plt.bar(x_axis, new_data_pyplot, alpha=.5, color="b", align="center")
plt.title("")
plt.xlabel("Drug Regimen")
plt.ylabel("Number of unique Mice Tested")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations,['Capomulin','Ceftamin','Infubinol','Ketapril','Naftisol','Placebo','Propriva','Ramicane','Stelasyn','Zoniferol'] , 
           rotation='vertical')
# Set x and y limits
plt.xlim(-0.75, len(x_axis))
plt.ylim(0, max(new_data_pyplot)+10)
plt.tight_layout()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pandas
plot_pie_pandas=merge_df.groupby(["Sex"]).count()["Mouse ID"]
plot_pie_pandas.plot(kind="pie", figsize=(7,4))
plt.tight_layout()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
plot_pie_pandas=merge_df.groupby(["Sex"]).count()["Mouse ID"]
Gender=["Female","Male"]
sizes=[plot_pie_pandas]
colors=["yellow","blue"]
explode=[0.1,0]
plt.pie(sizes, explode=explode, labels=Gender, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=180)
```
